```python
#Ejercicio 1

def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)

num = int(input("Ingrese un número: "))

for i in range(1, num + 1):
    print(f"Factorial de {i}: {factorial(i)}")
```

```python
#Ejercicio 2

def fibonacci(pos):
    if pos == 0:
        return 0
    elif pos == 1:
        return 1
    else:
        return fibonacci(pos - 1) + fibonacci(pos - 2)

num = int(input("Ingrese una posición: "))

for i in range(num + 1):
    print(fibonacci(i))
```

```python
#Ejercicio 3

def potencia(n, m):
    if m == 0:
        return 1
    else:
        return n * potencia(n, m - 1)

print(potencia(2, 5))
```

```python
#Ejercicio 4

def decimal_binario(n):
    if n == 0:
        return ""
    else:
        return decimal_binario(n // 2) + str(n % 2)

num = int(input("Ingrese un número decimal: "))

binario = decimal_binario(num)
print(binario if binario != "" else "0")
```

```python
#Ejercicio 5

def es_palindromo(palabra):
    if len(palabra) <= 1:
        return True
    if palabra[0] != palabra[-1]:
        return False
    return es_palindromo(palabra[1:-1])
```

```python
#Ejercicio 6

def suma_digitos(n):
    if n < 10:
        return n
    else:
        return (n % 10) + suma_digitos(n // 10)
```

```python
#Ejercicio 7

def contar_bloques(n):
    if n == 1:
        return 1
    else:
        return n + contar_bloques(n - 1)
```

```python
#Ejercicio 8

def contar_digito(numero, digito):
    if numero == 0:
        return 0
    ultimo = numero % 10
    if ultimo == digito:
        return 1 + contar_digito(numero // 10, digito)
    else:
        return contar_digito(numero // 10, digito)
```
