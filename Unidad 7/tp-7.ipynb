```python
#Ejercicio 5

frase = input("Ingrese una frase: ")

palabras_unicas = set(frase.split())
print("Las palabras únicas son:", palabras_unicas)

recuento_palabras = {palabra: frase.split().count(palabra) for palabra in palabras_unicas}
print("El recuento de palabras es:", recuento_palabras)
```

```python
#Ejercicio 6
alumnos = {}

for alumno in range(3):
    nombre = input(f"Ingrese el nombre del alumno {alumno+1}: ")
    lista_notas = []

    for nota in range(3):
        notas = float(input(f"Ingrese la calificación {nota+1} de {nombre}: "))
        lista_notas.append(notas)

    tuples_notas = tuple(lista_notas)
        
    alumnos[nombre] = tuples_notas

promedios = {}
for nombre, notas in alumnos.items():
    promedio = sum(notas) / len(notas)
    promedios[nombre] = promedio

print(alumnos)
print(promedios)
```

```python
#Ejercicio 7
aprobados_parcial_1 = []
aprobados_parcial_2 = []

for alumno in range(4):
    nombre = input(f"Ingrese el nombre del alumno {alumno+1} que aprobo el primer parcial: ")
    aprobados_parcial_1.append(nombre)

for alumno in range(4):
    nombre = input(f"Ingrese el nombre del alumno {alumno+1} que aprobo el segundo parcial: ")
    aprobados_parcial_2.append(nombre)

conjunto_parcial_1 = set(aprobados_parcial_1)
conjunto_parcial_2 = set(aprobados_parcial_2)

aprobados_ambos_parciales = conjunto_parcial_1 & conjunto_parcial_2
aprbaron_solo_un_parcial = conjunto_parcial_1 ^ conjunto_parcial_2
aprobaron_al_menos_un_parcial = conjunto_parcial_1 | conjunto_parcial_2

print("Alumnos que aprobaron ambos parciales:", aprobados_ambos_parciales)
print("Alumnos que aprobaron solo un parcial:", aprbaron_solo_un_parcial)
print("Alumnos que aprobaron al menos un parcial:", aprobaron_al_menos_un_parcial)

    
```

```python
#Ejercicio 8

ropa = {'remeras': 10, 'pantalones': 5, 'zapatos': 8, 'sombreros': 15}

while True:
    print("\n--- MENÚ DE STOCK ---")
    print("1. Consultar stock de un producto")
    print("2. Agregar unidades a un producto existente")
    print("3. Agregar un nuevo producto")
    print("4. Salir")

    opcion = input("Elegí una opción (1-4): ")

    if opcion == "1":
        producto = input("Ingresá el nombre del producto: ").lower()
        if producto in ropa:
            print("---------------------------------")
            print(f"El stock de {producto} es {ropa[producto]}")
            print("---------------------------------")
        else:
            print("---------------------------------")
            print("Ese producto no existe en el inventario.")
            print("---------------------------------")

    elif opcion == "2":
        producto = input("Ingresá el nombre del producto: ").lower()
        if producto in ropa:
            cantidad = int(input("¿Cuántas unidades querés agregar?: "))
            ropa[producto] += cantidad
            print("---------------------------------")
            print(f"Se agregaron {cantidad} unidades. Nuevo stock de {producto}: {ropa[producto]}")
            print("---------------------------------")
        else:
            print("---------------------------------")
            print("El producto no existe. Podés agregarlo desde la opción 3.")
            print("---------------------------------")

    elif opcion == "3":
        producto = input("Ingresá el nombre del nuevo producto: ").lower()
        if producto in ropa:
            print("---------------------------------")
            print("Ese producto ya existe.")
            print("---------------------------------")
        else:
            cantidad = int(input("Ingresá el stock inicial: "))
            ropa[producto] = cantidad
            print("---------------------------------")
            print(f"Producto {producto} agregado con {cantidad} unidades.")
            print("---------------------------------")

    elif opcion == "4":
        print("---------------------------------")
        print("Saliendo del sistema...")
        print("---------------------------------")
        break

    else:
        print("---------------------------------")
        print("Opción inválida. Intentá nuevamente.")
        print("---------------------------------")


```

```python
#Ejercicio 9 

agenda = {
    ("lunes", "10:00"): "Reunión",
    ("martes", "15:00"): "Clase de inglés",
    ("miércoles", "09:00"): "Gimnasio",
    ("jueves", "18:30"): "Programación 1",
    ("viernes", "20:00"): "Cena con amigos"
}

dia = input("Ingresá el día: ").lower()
hora = input("Ingresá la hora (por ejemplo 10:00): ")

clave = (dia, hora)

if clave in agenda:
    print(f"El evento del {dia} a las {hora} es: {agenda[clave]}")
else:
    print(f"No hay ningún evento registrado el {dia} a las {hora}.")
```

```python
#Ejercicio 10

original = {
    "Argentina": "Buenos Aires",
    "Chile": "Santiago",
    "Uruguay": "Montevideo",
    "Paraguay": "Asunción"
}

invertido = {}

for pais, capital in original.items():
    invertido[capital] = pais

print("Diccionario original:", original)
print("Diccionario invertido:", invertido)
```
